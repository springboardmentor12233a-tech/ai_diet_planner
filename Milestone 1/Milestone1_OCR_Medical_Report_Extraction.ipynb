OCR on all PNG images in /content folder and save to CSV

```python
!pip install easyocr opencv-python
import easyocr
import cv2
import os
import pandas as pd

reader = easyocr.Reader(['en'])

ocr_data = []

for file in os.listdir('/content'):
    if file.lower().endswith('.png'):
        img_path = os.path.join('/content', file)
        image = cv2.imread(img_path)

        results = reader.readtext(image)
        extracted_text = " ".join([text for _, text, _ in results])

        ocr_data.append({
            'Image_Name': file,
            'Extracted_Text': extracted_text
        })

ocr_df = pd.DataFrame(ocr_data)
ocr_df.head()
```

```python
# Save OCR results
ocr_df.to_csv('/content/OCR_Extracted_Reports.csv', index=False)

# Verify it worked by printing first few rows
print("OCR Data saved! Here are first few rows:\n")
print(ocr_df.head())
```

Upload a single image and perform OCR

```python
from google.colab import files

# Upload the file
uploaded = files.upload()  # This will open a file picker

# Get the filename
image_path = list(uploaded.keys())[0]

import easyocr
import cv2
reader = easyocr.Reader(['en'])

# Read the uploaded image
image = cv2.imread(image_path)

# Perform OCR
results = reader.readtext(image)

for bbox, text, confidence in results:
    print(f"{text} (Confidence: {confidence:.2f})")
```
