```python
image_path = r"C:\Users\localnaveen\Downloads\Picture-prescription1.jpg"

image = cv2.imread(image_path)

reader = easyocr.Reader(['en'])
results = reader.readtext(image)
```

```python
extracted_lines = []

for _, text, confidence in results:
    extracted_lines.append(text)

print("---- OCR EXTRACTED TEXT ----\n")
for i, line in enumerate(extracted_lines, start=1):
    print(f"{i}. {line}")
```

```python
combined_text = " ".join(extracted_lines).lower()

print("\n---- COMBINED TEXT USED FOR ANALYSIS ----\n")
print(combined_text)
```

```python
preferred_foods = []
avoid_foods = []

# Diabetes related
if "diabetes" in combined_text or "dm" in combined_text:
    preferred_foods.append("ü•ó Whole grains (brown rice, oats, millets)")
    preferred_foods.append("ü•ó High-fiber vegetables (beans, leafy greens)")
    preferred_foods.append("ü•ó Low-GI fruits (apple, guava, berries)")
    
    avoid_foods.append("‚ö†Ô∏è Sugar, sweets, chocolates")
    avoid_foods.append("‚ö†Ô∏è Sweetened drinks and packaged juices")
    avoid_foods.append("‚ö†Ô∏è Refined carbs (white bread, maida)")

# Hypertension / BP
if "hypertension" in combined_text or "bp" in combined_text:
    preferred_foods.append("ü•ó Low-salt foods")
    preferred_foods.append("ü•ó Potassium-rich foods (banana, spinach)")
    
    avoid_foods.append("‚ö†Ô∏è Excess salt")
    avoid_foods.append("‚ö†Ô∏è Pickles and processed foods")

# Lipid / Cholesterol
if "lipid" in combined_text or "dyslipidemia" in combined_text:
    preferred_foods.append("ü•ó Healthy fats (olive oil, nuts)")
    preferred_foods.append("ü•ó Omega-3 rich foods (fish, flax seeds)")
    
    avoid_foods.append("‚ö†Ô∏è Fried foods")
    avoid_foods.append("‚ö†Ô∏è Junk food and bakery items")

# Insulin / Blood sugar monitoring
if "insulin" in combined_text or "blood sugar" in combined_text:
    preferred_foods.append("ü•ó Fixed meal timings")
    preferred_foods.append("ü•ó Balanced meals with protein")
    
    avoid_foods.append("‚ö†Ô∏è Skipping meals")

# General recommendations
preferred_foods.append("ü•ó Drink enough water daily")
preferred_foods.append("ü•ó Eat fresh fruits and vegetables")

# ============================
# DISPLAY OUTPUT
# ============================

print("\nü•ó PREFERRED / RECOMMENDED FOODS\n")
for i, food in enumerate(preferred_foods, start=1):
    print(f"{i}. {food}")

print("\n‚ö†Ô∏è FOODS TO AVOID\n")
for i, food in enumerate(avoid_foods, start=1):
    print(f"{i}. {food}")
```
