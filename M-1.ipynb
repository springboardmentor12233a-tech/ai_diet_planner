```python
img = cv2.imread(image_path)

if img is None:
    print("âŒ Image not found. Check the file path.")
else:
    print("âœ… Image loaded successfully")

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.axis('off')
```
