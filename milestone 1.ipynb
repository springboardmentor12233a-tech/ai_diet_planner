```python
image = cv2.imread()
"PUN-0425-PA-0011414_Lab Report_27-04-2025_0949-24_PM@F.pdf_page_10.png"
```

```python
reader = easyocr.Reader(['en'])
results = reader.readtext(image)
```

```python
print("Extracted Text:\n")
for bbox, text, confidence in results:
    print(f"{text} (Confidence: {confidence:.2f})")
```

```python
ocr_text = " ".join([text for _, text, _ in results])
ocr_text
```

```python
data = {}

hb = re.search(r'HEMOGLOBIN\s*[:\-]?\s*([\d.]+)', ocr_text, re.I)
data['Hemoglobin'] = hb.group(1) if hb else None

rbc = re.search(r'R\.?B\.?C\.?\s*[:\-]?\s*([\d.]+)', ocr_text, re.I)
data['RBC_Count'] = rbc.group(1) if rbc else None

wbc = re.search(r'W\.?B\.?C\.?\s*[:\-]?\s*(\d+)', ocr_text, re.I)
data['WBC_Count'] = wbc.group(1) if wbc else None

platelets = re.search(r'PLATELET\s*COUNT\s*[:\-]?\s*([\d.]+)', ocr_text, re.I)
data['Platelet_Count'] = platelets.group(1) if platelets else None

data
```

```python
structured_df = pd.DataFrame([data])
structured_df
```
