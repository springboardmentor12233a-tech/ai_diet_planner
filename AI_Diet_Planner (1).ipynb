```python
import easyocr
import cv2

image_path = "/content/AHD-0425-PA-0007561_JITENDRA TRIVEDI DS_28-04-2025_1019-21_AM.pdf_page_9.png"

image = cv2.imread(image_path)

reader = easyocr.Reader(['en'])
results = reader.readtext(image)

print("Extracted Text from Scanned Medical Report:\n")
for bbox, text, confidence in results:
    print(f"{text} (Confidence: {confidence:.2f})")
```
