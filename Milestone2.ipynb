```python
import pytesseract
from PIL import Image
import os

image_dir = "/content/drive/MyDrive/dataset/images"
ocr_texts = []

for root, dirs, files in os.walk(image_dir):
    for file in files:
        if file.lower().endswith((".png", ".jpg", ".jpeg")):
            img_path = os.path.join(root, file)
            img = Image.open(img_path)
            text = pytesseract.image_to_string(img)
            ocr_texts.append({
                "image_name": file,
                "extracted_text": text
            })
```

```python
ocr_df = pd.DataFrame(ocr_texts)

ocr_df.to_csv(
    "/content/drive/MyDrive/dataset/output/ocr_extracted_text.csv",
    index=False
)
```

```python
def check_thresholds(patient):
    alerts = []

    if patient["Glucose"] > 140:
        alerts.append("High Glucose Level")

    if patient["BloodPressure"] > 90:
        alerts.append("High Blood Pressure")

    if patient["BMI"] > 30:
        alerts.append("High BMI (Obesity Risk)")

    if patient["Insulin"] > 200:
        alerts.append("Abnormal Insulin Level")

    return alerts
```

```python
def health_analysis(input_data):
    columns = X.columns
    patient = dict(zip(columns, input_data))

    # ML Prediction
    prediction = model.predict([input_data])[0]
    status = "CRITICAL (Diabetic)" if prediction == 1 else "NORMAL"

    # Threshold Alerts
    alerts = check_thresholds(patient)

    return {
        "Health Status": status,
        "Alerts": alerts
    }
```
